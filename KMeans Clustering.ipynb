## Let's Begin

### Import necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

### Read 'Credit Card Customer Data.csv' and store it in a DataFrame variable

```python
df = pd.read_csv('Credit Card Customer Data.csv')
```

### View top 5 rows

```python
df.head()
```

### View info

```python
df.info()
```

### View statistical information about the dataset

```python
df.describe()
```

### Check for any null values

```python
df.isna().sum()
```

### Check the shape of the dataset

```python
df.shape
```

### Drop irrelevent columns (SI_No and Customer Key)

```python
df.drop(columns = ['Sl_No', 'Customer Key'], inplace = True)
df.head()
```

## <span style = "color:orange"> Visualization </span>

### Plot a scatterplot of Avg_credit_Limit vs Total_Credit_Cards with respect to Total_calls_made

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', hue = 'Total_calls_made', data = df, palette = 'rainbow')
plt.show()
```

### Plot correlation matix on Heatmap

```python
plt.figure(figsize = (8,6))
sns.heatmap(df.corr(),annot = True, cmap = 'coolwarm')
plt.show()
```

### Plot a pairplot

```python
sns.pairplot(df)
plt.show()
```

### Create K Means model (Let k = 4)

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4)
```

### Fit predict for our Dataset

```python
clusters = model.fit_predict(df)
clusters
```

### Assign the predicted clusters to a new column in our dataset

```python
df['clusters'] = clusters
```

```python
df.head()
```

## Visualize the clusters

```python
# Use x = 'Avg_Credit_Limit' ang y =  'Total_Credit_Cards'
df['clusters'].unique()
```

```python
df0 = df[df['clusters'] == 0]
df1 = df[df['clusters'] == 1]
df2 = df[df['clusters'] == 2]
df3 = df[df['clusters'] == 3]
```

```python
sns.scatterplot(x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', data = df0, color = 'red')
sns.scatterplot(x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', data = df1, color = 'green')
sns.scatterplot(x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', data = df2, color = 'blue')
sns.scatterplot(x = 'Avg_Credit_Limit', y = 'Total_Credit_Cards', data = df3, color = 'violet')

plt.show()
```
